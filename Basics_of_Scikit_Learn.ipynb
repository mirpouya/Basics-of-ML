```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
# creating a data set 

N = 50
x = np.arange(N)
y = 4*x - 7 + (15*np.random.randn(N))

plt.scatter(x,y)
```

```python
# importing sklearn

from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)

# fitting regression line
model.fit(x[:, np.newaxis], y)   # x.shape = (N,), and .fit(.) funciton cannot accept this shape of input. Therefore we add np.newaxis

# storing predicitons of the model
y_pred = model.predict(x[:, np.newaxis])
plt.scatter(x, y)
plt.plot(x, y_pred, c="red")
```

<h2> Generate synthetic data </h2>

```python
from sklearn.datasets import make_blobs
```

```python
# creating fake dataset

X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.2)
plt.scatter(X[:,0], X[:,1], c=y, cmap="autumn")
```

```python
# spliting the synthetic data into train and test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
# import svm classifier

from sklearn.svm import SVC
```

```python
model = SVC(gamma="auto")
model.fit(X_train, y_train)
```

```python
# evaluating prediction

y_pred = model.predict(X_test)

acc = 100 * np.sum(y_pred == y_test)/y_test.size
print("accuracy is {}%".format(acc))
```

```python
# Random Forest clasifier

from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
acc = 100*np.sum(y_pred == y_test)/y_test.size
print("Accuracy is {}%".format(acc))
```
